## **SPEAKER RECOGNITION SYSTEM**

```python
import os
import random
from glob import glob
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import librosa
import librosa.display
import math

from sklearn.preprocessing import LabelEncoder, StandardScaler, label_binarize
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve, auc, precision_score,
                             recall_score, f1_score)

import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (Conv1D, BatchNormalization, MaxPooling1D,
                                     Dropout, LSTM, Dense, GlobalAveragePooling1D,
                                     Flatten)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

DATASET_PATH = "/kaggle/input/speaker-recognition-dataset/16000_pcm_speeches/"
SPEAKERS = ['Nelson_Mandela', 'Benjamin_Netanyau', 'Magaret_Tarcher', 'Jens_Stoltenberg', 'Julia_Gillard']  # exactly these
BACKGROUND_FOLDER_NAME = "_background_noise_"
SAMPLE_RATE = 16000
N_MFCC = 13
MAX_TIME_STEPS = 100        # time-steps after padding/truncating (frames)
NOISE_FACTOR = 0.4         
SPEC_AUG = True            
RANDOM_SEED = 42
BATCH_SIZE = 16
EPOCHS = 30
DROPOUT = 0.3

random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

print("DATASET_PATH:", DATASET_PATH)
found = []
for entry in sorted(os.listdir(DATASET_PATH)):
    full = os.path.join(DATASET_PATH, entry)
    if os.path.isdir(full) and entry in SPEAKERS:
        found.append(entry)
print("Speakers found (will use):", found)
assert set(found) == set(SPEAKERS), "Make sure those speaker folders exist in DATASET_PATH"

# background noises
bg_folder = os.path.join(DATASET_PATH, BACKGROUND_FOLDER_NAME)
bg_noise_files = []
if os.path.isdir(bg_folder):
    bg_noise_files = [os.path.join(bg_folder, f) for f in os.listdir(bg_folder) if f.endswith('.wav')]
print("Background noise files:", len(bg_noise_files))

# -----------------------------
# Utilities: safe time-stretch (phase vocoder) and augmentations
# -----------------------------
def safe_time_stretch(y, rate):
    """
    Time-stretch waveform using STFT + phase_vocoder to avoid librosa API differences.
    Clips/pads to original length.
    """
    try:
        # compute STFT
        hop_length = 256
        n_fft = 1024
        D = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
        D_stretched = librosa.phase_vocoder(D, rate, hop_length=hop_length)
        y_stretch = librosa.istft(D_stretched, hop_length=hop_length, length=len(y))
        return y_stretch
    except Exception as e:
        # fallback: naive resample (changes pitch but better than crash)
        try:
            sr_in = SAMPLE_RATE
            target_sr = int(sr_in / rate)
            y_rs = librosa.resample(y, orig_sr=sr_in, target_sr=target_sr)
            y_back = librosa.resample(y_rs, orig_sr=target_sr, target_sr=sr_in)
            if len(y_back) > len(y):
                y_back = y_back[:len(y)]
            else:
                y_back = np.pad(y_back, (0, max(0, len(y)-len(y_back))))
            return y_back
        except Exception:
            return y  

def add_background_noise(y):
    if not bg_noise_files:
        return y
    noise_path = random.choice(bg_noise_files)
    noise, _ = librosa.load(noise_path, sr=SAMPLE_RATE)
    if len(noise) < len(y):
        reps = int(np.ceil(len(y) / len(noise)))
        noise = np.tile(noise, reps)[:len(y)]
    else:
        noise = noise[:len(y)]
    return y + NOISE_FACTOR * noise

def pitch_shift_waveform(y, sr):
    n_steps = random.uniform(-2.0, 2.0)
    try:
        return librosa.effects.pitch_shift(y, sr=sr, n_steps=n_steps)
    except Exception:
        # fallback: naive resample that approximates pitch shift
        rate = 2.0 ** (n_steps / 12.0)
        y_rs = librosa.resample(y, orig_sr=sr, target_sr=int(sr * rate))
        y_back = librosa.resample(y_rs, orig_sr=int(sr * rate), target_sr=sr)
        if len(y_back) > len(y):
            y_back = y_back[:len(y)]
        else:
            y_back = np.pad(y_back, (0, len(y)-len(y_back)))
        return y_back

def random_augment_waveform(y, sr):
    out = np.copy(y)
    # Add background noise with prob 0.6
    if bg_noise_files and random.random() < 0.6:
        out = add_background_noise(out)
    # Pitch shift with prob 0.5
    if random.random() < 0.5:
        out = pitch_shift_waveform(out, sr)
    # Time stretch with prob 0.5
    if random.random() < 0.5:
        rate = random.uniform(0.85, 1.15)
        out = safe_time_stretch(out, rate)
    # Clip to [-1,1]
    out = out / np.max(np.abs(out)) if np.max(np.abs(out))>0 else out
    return out

# -----------------------------
# Feature extraction (MFCC + delta + delta2) 
# -----------------------------
def extract_mfcc_features(file_path, augment=False, spec_aug=SPEC_AUG):
    y, sr = librosa.load(file_path, sr=SAMPLE_RATE)
    # optionally augment waveform
    if augment:
        y = random_augment_waveform(y, sr)
    # compute MFCCs
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC)
    delta = librosa.feature.delta(mfcc)
    delta2 = librosa.feature.delta(mfcc, order=2)
    combined = np.vstack([mfcc, delta, delta2])  # shape (39, T)
    combined = combined.T  # (T, 39)
    # spec augment (simple masking)
    if augment and spec_aug:
        t, f = combined.shape
        # time mask
        t_mask = random.randint(0, max(0, t // 8))
        if t_mask > 0:
            t0 = random.randint(0, max(0, t - t_mask))
            combined[t0:t0+t_mask, :] = 0
        # freq mask
        f_mask = random.randint(0, max(0, f // 8))
        if f_mask > 0:
            f0 = random.randint(0, max(0, f - f_mask))
            combined[:, f0:f0+f_mask] = 0
    # pad/truncate to MAX_TIME_STEPS
    if combined.shape[0] > MAX_TIME_STEPS:
        combined = combined[:MAX_TIME_STEPS, :]
    else:
        pad = MAX_TIME_STEPS - combined.shape[0]
        combined = np.pad(combined, ((0, pad), (0, 0)), mode='constant')
    return combined, y

all_files = []
for spk in SPEAKERS:
    spk_folder = os.path.join(DATASET_PATH, spk)
    if not os.path.isdir(spk_folder):
        raise FileNotFoundError(f"Speaker folder not found: {spk_folder}")
    wavs = [f for f in sorted(os.listdir(spk_folder)) if f.endswith('.wav')]
    for w in wavs:
        all_files.append((spk, os.path.join(spk_folder, w)))

print("Total WAV files found (across speakers):", len(all_files))

X = []
y_labels = []
file_paths = [] 

for spk, fp in tqdm(all_files, desc="Processing all audio files"):
    feat, orig_y = extract_mfcc_features(fp, augment=False, spec_aug=False)
    X.append(feat)
    y_labels.append(spk)
    file_paths.append(fp)
    feat_aug, _ = extract_mfcc_features(fp, augment=True, spec_aug=SPEC_AUG)
    X.append(feat_aug)
    y_labels.append(spk)
    file_paths.append(fp + "||aug")

X = np.array(X)   
y_labels = np.array(y_labels)
print("Features array shape:", X.shape)
print("Labels shape:", y_labels.shape)

def plot_waveform_and_mfcc(fp, augment=False):
    feat, orig = extract_mfcc_features(fp, augment=augment, spec_aug=False)  
    y, sr = librosa.load(fp, sr=SAMPLE_RATE)
    if augment:
        y = random_augment_waveform(y, sr)
    plt.figure(figsize=(12, 5))
    plt.subplot(1,2,1)
    plt.plot(y)
    plt.title(f"Waveform ({'aug' if augment else 'orig'}) - {os.path.basename(fp)}")
    plt.subplot(1,2,2)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=N_MFCC)
    librosa.display.specshow(mfcc, x_axis='time', sr=sr)
    plt.title(f"MFCC ({'aug' if augment else 'orig'})")
    plt.colorbar()
    plt.tight_layout()
    plt.show()

sample_examples = random.sample(all_files, 2)
for spk, fp in sample_examples:
    print("Sample:", spk, fp)
    plot_waveform_and_mfcc(fp, augment=False)
    plot_waveform_and_mfcc(fp, augment=True)

le = LabelEncoder()
y_encoded = le.fit_transform(y_labels)
num_classes = len(le.classes_)
print("Classes:", le.classes_, "Num classes:", num_classes)

indices = np.arange(len(X))
np.random.shuffle(indices)
X = X[indices]
y_encoded = y_encoded[indices]
file_paths = [file_paths[i] for i in indices]

X_train, X_test, y_train, y_test, paths_train, paths_test = train_test_split(
    X, y_encoded, file_paths, test_size=0.2, random_state=RANDOM_SEED, stratify=y_encoded
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

nsamples, tsteps, nfeat = X_train.shape
scaler = StandardScaler()
X_train_flat = X_train.reshape(-1, nfeat)
X_train_flat = scaler.fit_transform(X_train_flat)
X_train = X_train_flat.reshape(-1, tsteps, nfeat)

X_test_flat = X_test.reshape(-1, nfeat)
X_test_flat = scaler.transform(X_test_flat)
X_test = X_test_flat.reshape(-1, tsteps, nfeat)

print("After scaling -> Train shape:", X_train.shape, "Test shape:", X_test.shape)

input_shape = (X_train.shape[1], X_train.shape[2]) 
model = Sequential([
    Conv1D(64, 3, padding='same', activation='relu', input_shape=input_shape),
    BatchNormalization(),
    Conv1D(64, 3, padding='same', activation='relu'),
    MaxPooling1D(2),
    Dropout(DROPOUT),

    Conv1D(128, 3, padding='same', activation='relu'),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(DROPOUT),

    LSTM(64, return_sequences=False),
    Dropout(DROPOUT),

    Dense(128, activation='relu'),
    Dropout(DROPOUT),
    Dense(num_classes, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[early_stop, reduce_lr],
    verbose=1
)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
plt.tight_layout(); plt.show()

y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix"); plt.show()

y_test_bin = label_binarize(y_test, classes=np.arange(num_classes))
roc_auc_overall = roc_auc_score(y_test_bin, y_pred_prob, multi_class='ovr')
print(f"Overall ROC-AUC: {roc_auc_overall:.6f}")

metrics_list = []
for i, spk in enumerate(le.classes_):
    y_true_i = y_test_bin[:, i]
    y_pred_label_i = (y_pred == i).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true_i, y_pred_label_i).ravel()
    acc = (tp + tn) / (tp + tn + fp + fn)
    prec = precision_score(y_true_i, y_pred_label_i, zero_division=0)
    rec = recall_score(y_true_i, y_pred_label_i, zero_division=0)
    spec = tn / (tn + fp) if (tn + fp) > 0 else 0.0
    f1s = f1_score(y_true_i, y_pred_label_i, zero_division=0)
    roc_i = roc_auc_score(y_true_i, y_pred_prob[:, i])
    metrics_list.append({
        'Speaker': spk,
        'Accuracy': f"{acc:.4f}",
        'Precision': f"{prec:.4f}",
        'Recall': f"{rec:.4f}",
        'Specificity': f"{spec:.4f}",
        'F1-score': f"{f1s:.4f}",
        'ROC-AUC': f"{roc_i:.6f}"
    })
import pandas as pd
metrics_df = pd.DataFrame(metrics_list).set_index('Speaker')
print("\nPer-class metrics:")
print(metrics_df)

# ROC curves plot
plt.figure(figsize=(9,7))
colors = ['blue','red','green','orange','purple'][:num_classes]
for i, color in enumerate(colors):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])
    auc_i = auc(fpr, tpr)
    plt.plot(fpr, tpr, color=color, lw=2, label=f"{le.classes_[i]} (AUC={auc_i:.6f})")
plt.plot([0,1],[0,1],'k--', lw=1)
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curves per class"); plt.legend(loc='lower right'); plt.grid(True)
plt.show()

random.seed(RANDOM_SEED)
fig = plt.figure(figsize=(18, 12))
plot_idx = 1
for i, spk in enumerate(le.classes_):
    indices = np.where(y_test == i)[0]
    chosen = random.sample(list(indices), k=2)
    for idx in chosen:
        x_sample = X_test[idx]  # already scaled MFCC features
        # For plotting MFCC we want to invert scaling per feature for readability (approx)
        # We'll just show the feature matrix as-is (scaled) but titles indicate true/pred
        pred_label = le.inverse_transform([y_pred[idx]])[0]
        true_label = le.inverse_transform([y_test[idx]])[0]
        ax = plt.subplot(5, 2, plot_idx)
        plt.imshow(x_sample.T, aspect='auto', origin='lower')
        plt.title(f"True: {true_label} | Pred: {pred_label}")
        plt.xlabel("Time steps"); plt.ylabel("MFCC+delta features")
        plot_idx += 1
plt.tight_layout(); plt.show()

rand10 = random.sample(range(len(X_test)), 10)
print("\nIndex  True -> Predicted")
for idx in rand10:
    print(f"{idx:4d}  {le.inverse_transform([y_test[idx]])[0]:20s} -> {le.inverse_transform([y_pred[idx]])[0]:20s}")

test_loss, test_acc = model.evaluate(X_test, y_test, verbose=1)
print("\nFinal test accuracy:", test_acc, "test loss:", test_loss)
```
